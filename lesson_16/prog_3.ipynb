```python
import numpy as np
import matplotlib.pyplot as plt

# this block approximates with function of polynomial power of 1
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)

x = np.linspace(1, 15, 100)
y = np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)

b1 = [f(i) for i in np.array([1, 15])]

A1 = np.array([[1, 1], [1, 15]])
w1 = np.linalg.solve(A1, b1)

y1 = sum([w1[i]*x**i for i in range(len(w1))])
```

```python
# this block approximates with function of polynomial power of 2
b2 = [f(i) for i in np.array([1, 8, 15])]

A2 = np.array([[1, 1, 1], [1, 8, 8**2], [1, 15, 15**2]])
w2 = np.linalg.solve(A2, b2)

y2 = sum([w2[i]*(x**i) for i in range(len(w2))])
```

```python
# this block approximates with function of polynomial power of n - 1 (you can enter any power)
power = 3
n = power + 1

x0 = np.linspace(1, 15, n)
b0 = np.array([f(item) for item in x0])

A0 = np.ones((n, n))
for i in range(n):
    for j in range(n):
        A0[i][j] = (x0[i]**j)

w0 = np.linalg.solve(A0, b0)
y0 = sum([w0[i]*(x**i) for i in range(len(w0))])
```

```python
plt.plot(x, y, 'r-') # real function
plt.plot(x, y0, 'y-') # approximates with function of polynomial power of n - 1
plt.plot(x, y1, 'g-') # approximates with function of polynomial power of 1
plt.plot(x, y2, 'b-') # approximates with function of polynomial power of 2
plt.legend(np.array(['real', 'power of n-1', 'power of 1', 'power of 2']))

plt.grid()
plt.show()
```
